```python
def clean_words(word, township = False):
    #if type(word) in ['str', 'unicode']:
    if word is not None:
        if township:
            word = word.split('(')[0]
         #checking non-english characters
        word = " ".join(re.findall("[\x00-\x7F]+", word))
    #word = word.replace('(', ' ')
    #print word.split('(')[0]
        if word.strip() == '':
            word = None
    return word
```
